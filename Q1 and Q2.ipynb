```python
class Data:
  def __init__(self, name, projection) -> None:
      self.name = name
      self.dim = projection
      self.values = None

  def get_size(self, dim_values):
      flattened_array = self.dim.flatten()
      projection = [dim_values[dim] for dim in flattened_array]
      inp = 1 if self.name == "Inputs" else 0
      shape = np.array(projection).reshape(self.dim.shape).sum(axis=1).squeeze(axis=-1) - inp
      return shape

  def init_values(self, shape):
    if self.name == "Outputs":
      self.values = np.zeros(shape=shape)
      self.values[:] = np.nan
    elif self.name == "Inputs":
      self.values = np.ones(shape=shape)
    else:
      self.values = np.arange(np.array(shape).prod()).reshape(shape)
```

```python
class Problem:
  def __init__(self, name, dims, instance) -> None:
      self.name = name
      self.dims = dims
      self.instance = instance

  @classmethod
  def read_yaml(cls, path):
    with open(path) as stream:
      data = yaml.safe_load(stream)
    name = data['problem']['shape']['name']
    dims = [data['problem']['instance'][dim] for dim in data['problem']['shape']['dimensions']]
    problem = cls(name, dims, data['problem']['instance'])
    for data_space in data['problem']['shape']['data-spaces']:
      data_obj_name = data_space['name']
      data_obj_projection_symb = np.array(data_space['projection'])
      shape = data_obj_projection_symb.shape
      data_obj_projection = data_obj_projection_symb#.reshape(*list(shape)[:-1])
      # flattened_array = data_obj_projection_symb.flatten()
      # data_obj_projection = [data['problem']['instance'][dim] for dim in flattened_array]
      # data_obj_projection = np.array(data_obj_projection).reshape(shape).sum(axis=1).squeeze()
      data_obj = Data(data_obj_name, data_obj_projection)
      setattr(problem, data_obj_name, data_obj)

    return problem
```

```python
class Memory:
    def __init__(self, name, depth, width, block_size, word_bits) -> None:
        self.name = name
        self.depth = depth
        self.width = width
        self.block_size = block_size
        self.word_bits = word_bits
        self.size_dict = None
        self.content_I = []
        self.content_W = []
        self.content_O = []
        self.parent = None
        self.n_writes = {'Outputs':0, 'Inputs':0, 'Weights':0}
        self.n_reads = {'Outputs':0, 'Inputs':0, 'Weights':0}
        self.n_updates = {'Outputs':0, 'Inputs':0, 'Weights':0}

    def read(self, address, type):
      content = self.get_container(type)
      value = list(filter(lambda val: val['address'] == address and val['type']==type, content))
      if value:
        value = value[0]
      elif self.parent is not None:
        value = self.parent.read(address, type)

        self.write(value)
        if (type=='Outputs') and (np.isnan(value['value'])):
          self.parent.n_reads[type] -=1
          # self.n_writes[type] -=1

          value = {'address':address, 'type':type, 'value':np.nan}
          # content.append(value)

      else:
        value = None

      if value:
        self.n_reads[type] += 1
      return value

    def add_parent(self, parent):
      self.parent = parent

    def write_back(self, value):
      self.parent.write(value)

    def flush_all(self, type_=None):
      contents = [self.content_I, self.content_O, self.content_W]
      types = ['Inputs', 'Outputs', 'Weights']
      for type,content in zip(types,contents):
        if type_ is None or type_==type:
          for i in range(len(content)):
            self.flush(type)

    def flush(self,type):
      content = self.get_container(type)
      value = content[0]
      value = content.pop(0)
      if value['type'] in ['Outputs']:
        self.n_reads[value['type']] += 1
        self.write_back(value)


    def get_container(self, type):
      if type=='Outputs':
        return self.content_O
      elif type=='Inputs':
        return self.content_I
      else:
        return self.content_W

    def get_depth(self,type):
      return self.size_dict[type]

    def write(self, val):

      if val and not (np.isnan(val['value'])):
        content = self.get_container(val['type'])
        self.n_writes[val['type']] += 1
        if len(content) >= self.get_depth(val['type']):

          self.flush(val['type'])
        if content:
          for i, ele in enumerate(content):
            if ele['address'] == val['address'] and ele['type'] == val['type']:
              content[i] = val.copy()
              self.n_writes[val['type']] -= 1
              self.n_updates[val['type']] += 1
              return
        content.append(val.copy())

    def __repr__(self):
        return f"{self.name}(depth:{self.depth})"
```

```python
class intmac:
  def __init__(self, name, datawidth):
      self.name = name
      self.datawidth = datawidth

  def add_parent(self, parent):
      self.parent = parent

  def read(self, address, type):
      return self.parent.read(address, type)

  def write(self, val):
      self.parent.write(val)
  def __repr__(self):
        return f"{self.name}(datawidth:{self.datawidth})"
```

```python
class PE:
  def __init__(self, name, local) -> None:
     self.name = name
     self.local = local
     self.parent = None

  def add_parent(self, parent):
      self.parent = parent

  def __repr__(self):
        return f"{self.name}(local:{self.local})"
```

```python
class System:
  def __init__(self, name, technology, local, subtree) -> None:
      self.name = name
      self.technology = technology
      self.local = local
      self.subtree = subtree
      self.parent = None

  def add_parent(self, parent):
      self.parent = parent

  def __repr__(self):
      return f"""{self.name}(local:{self.local},
            subtree:
            {self.subtree})"""


```

```python
def read_sys(sys_info):
    sys_tech = sys_info.get('technology', None)
    sys_name = sys_info['name']
    sys_local = []
    for c in sys_info['local']:
      c_name = c['name']
      c_attr = c['attributes']
      c_attr = {k.replace('-', '_'):v for k, v in c_attr.items()}
      c_attr['name'] = c_name
      c_class = eval(c['class'])
      obj = c_class(**c_attr)
      sys_local.append(obj)
    sys_attr = sys_info.get('attributes', {})
    sys_attr['name'] = sys_name
    sys_attr['local'] = sys_local
    return sys_attr
```

```python
class Arch:
  def __init__(self, version, subtree) -> None:
     self.version = version
     self.subtree = subtree
     self.memory_levels = {}
     self.macs = []
     self.max_memory_level = None
     self.add_levels()
     self.add_mac()

  def add_levels_recursive(self, subtree, level=0):
      levels = {}

      for system in subtree:
          local = [element for element in getattr(system, 'local', []) if isinstance(element, Memory)]
          if local:
              levels.setdefault(level, []).extend(local)
          sublevels = self.add_levels_recursive(getattr(system, 'subtree', []), level + 1)
          for l, memories in sublevels.items():
              for memory in memories:
                if local:
                  memory.add_parent(local[0])
              levels.setdefault(l, []).extend(memories)

      return levels

  def add_levels(self):
      levels = self.add_levels_recursive(self.subtree)
      self.memory_levels ={len(levels)-k-1:v for k, v in levels.items()}
      self.max_memory_level = self.memory_levels.get(max(self.memory_levels))

  def add_mac(self):
    macs = []
    for system in self.subtree:
      for pe in system.subtree:
        for ele in pe.local:
          if isinstance(ele, intmac):
            macs.append(ele)
    self.macs = macs

  @classmethod
  def read_yaml(cls, path):
    with open(path) as stream:
      data = yaml.safe_load(stream)
    version = data['architecture']['version']
    subtree = []
    for sys_info in data['architecture']['subtree']:
      sys_attr = read_sys(sys_info)
      sys_subtree = []
      for c in sys_info['subtree']:
        c_attr = read_sys(c)
        pe = PE(**c_attr)
        #add buffer as parent for macc
        pe.local[1].add_parent(pe.local[0])
        sys_subtree.append(PE(**c_attr))
      sys_attr['subtree'] = sys_subtree
      subtree.append(System(**sys_attr))
    obj = cls(version, subtree)
    return obj


```

```python
def get_vectors(m):
    shape = m.problem.Outputs.get_size(m.problem.instance)
    o = np.ones(shape)
    for d in filter(lambda x:x['type']=='Outputs',m.arch.max_memory_level[0].content_O):
      idx = d['address']
      val = d['value']
      o.__setitem__(idx,val)

    shape = m.problem.Inputs.get_size(m.problem.instance)
    I = np.ones(shape)
    for d in filter(lambda x:x['type']=='Intputs',m.arch.max_memory_level[0].content_I):
      idx = d['address']
      val = d['value']
      I.__setitem__(idx,val)


    shape = m.problem.Weights.get_size(m.problem.instance)
    W = np.ones(shape)
    for d in filter(lambda x:x['type']=='Weights',m.arch.max_memory_level[0].content_W):
      idx = d['address']
      val = d['value']
      W.__setitem__(idx,val)

    return o, I, W

import torch.functional as F
def convolution_1d_multichannel(weights, input):
    num_channels = len(weights)
    weight_length = len(weights[0])
    result = np.zeros((num_channels, len(input)-weight_length+1))

    for i in range(len(input)-weight_length+1):
        for c in range(num_channels):
            start_idx = i
            end_idx = i + weight_length
            result[c][i] += np.sum(weights[c] * input[start_idx:end_idx])

    return result


def check(m):
  o, I, W = get_vectors(m)
  result = convolution_1d_multichannel(W, I)
  return (o!=result).sum()
```

```python
class LevelMap:
  def __init__(self, target, type, factors, permutation, level):
    self.target = target
    self.type = type
    self.factors = factors
    self.permutation = permutation
    self.level = level

  def __repr__(self) -> str:
      return f"LevelMap({self.target}, {self.type}, {self.factors}, {self.permutation}, {self.level})"
```

```python
def check_equal_elements(list1, list2):
    # Check if the lengths of the lists are equal
    if len(list1) != len(list2):
        return False

    # Iterate through all indices and compare dictionaries
    for index in range(len(list1)):
        dict1 = list1[index]
        dict2 = list2[index]
        if dict1 != dict2:
            return False

    return True
```

```python
def YAML_generator(component_class="SRAM", depth=64,
                   width=8, path="common/Q2/arch_base.yaml", f=""):
    current_path = os.getcwd()
    path = os.path.join(current_path, path)
    with open(path, 'r') as file:
        txt = file.read()
        txt_new = txt.replace('$depth$', str(depth))
        txt_new = txt_new.replace('$width$', str(width))
        txt_new = txt_new.replace('$c_class$', component_class)
        with open(f, 'w') as output:
            output.write(txt_new)
```

```python
def YAML_parser(path="timeloop-model.ERT_summary.yaml"):
    current_path = os.getcwd()
    path = os.path.join(current_path, path)
    with open(path, 'r') as stream:
        data_loaded = yaml.safe_load(stream)
    for data in data_loaded['ERT_summary']['table_summary'][0]['actions']:
        if data['name'] == 'read':
            if "max_energy" in data.keys():
                energy_read = data['max_energy']
            else :
                energy_read = data['energy']
        if data['name'] == 'write':
            if "max_energy" in data.keys():
                energy_write = data['max_energy']
            else :
                energy_write = data["energy"]
    return energy_read, energy_write
```

```python
c_class_s = ["DRAM", "regfile", "SRAM"]
c_dep_wid = list(product(["DRAM", "regfile", "SRAM"],[8192],[16, 128,1024,2048])) + list(product(["DRAM", "regfile", "SRAM"],[512,2048,4096, 4608,5632,6656,8192],[128]))
print(len(c_dep_wid),c_dep_wid)
```
